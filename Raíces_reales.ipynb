```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
def ecuacion ():
    discriminante=0
    x = random.uniform(0, 2)
    y = random.uniform(0, 2)
    if (x**2-4*y)>=0:
        discriminante += 1
    return discriminante
    print(discriminante)
```

```python
ecuacion ()
```

```python
def prob_real(n):
    prob = []
    inter = 0
    for i in range(n):
        inter += ecuacion()
        prob.append(inter/(i+1))
    return prob
```

```python
X = prob_real(1000)
print(np.mean(X), np.std(X))
```

```python
plt.plot(X)
```
